Consider a (univariate) lognormal random variable: $X\sim \log N(\mu,\sigma^2)$.

a) Write a Python function that determines the value of the parameters $\mu$ and $\sigma^2$ from the expectation $E\{X\}$ and the variance $V\{X\}$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mu_sigma2(E,Var):
    mu = np.log(E) - 0.5*Var
    sigma2=np.log(1+Var/E**2)
    return (mu, sigma2)
```

b) Write a Python script in which you:

-- Use the function created in a) to determine the parameters $\mu$ and $\sigma^2$ such that $E\{X\}=3$  and $V\{X\}=5$.

-- Generate a large sample from the distribution of $X$ with such parameters;

-- Plot the sample (do not join the observations, show them as dots);

-- Plot the histogram (suitably normalized) and superimpose the exact pdf for comparison;

-- Plot the empirical cdf and superimpose the exact cdf for comparison.

Note: display title, labels and legend in each plot.

```python
mu,sigma2=mu_sigma2(3, 5)
mu,sigma2
```

```python
sample=np.random.lognormal(mu,np.sqrt(sigma2),1000)
sample[21:50]
```

```python
plt.plot(sample,'ob')
plt.title('Sample of lognormal distribution with E=3 and Var=5')
plt.xlabel('the ranks of the realisations')
plt.ylabel('values of the realisations')
plt.show()
```

```python
count, bins, ignored = plt.hist(sample, 100, normed=True, align='mid')
plt.title('histogram')
plt.xlabel('values of realisations')
plt.ylabel('frequences normalized')
```

```python
x = np.linspace(min(bins), max(bins), 10000)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma2))/(x * np.sqrt(sigma2) * np.sqrt(2 * np.pi)))
```

```python
plt.hist(sample, 100, normed=True, align='mid')
plt.plot(x, pdf, linewidth=2, color='r')
plt.title('histogram and the exact pdf')
plt.xlabel('values of realizations')
plt.ylabel('pdf')
plt.show()
```
